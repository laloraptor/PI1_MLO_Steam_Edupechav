## **1. Función PlayTimeGenre (FUNCIONA. Revisado)**

PlayTimeGenre

Para esta función, necesitarás realizar un merge entre df_juegos y df_usit para combinar la información de género (genres) y horas jugadas (playtime_forever). El objetivo es calcular las horas jugadas por género para cada año y luego identificar el año con más horas jugadas para el género dado.

### **1. Codigo de la función playtime_genre**

```python

def playtime_genre(genero: str):
    # Filtra el DataFrame por el género específico
    filtered_df = df_f1[df_f1['genres'] == genero]

    # Agrupa por año y calcula las horas jugadas sumando
    grouped_df = filtered_df.groupby('year_release')['playtime_forever'].sum().reset_index()

    # Encuentra el año con más horas jugadas
    max_year = grouped_df.loc[grouped_df['playtime_forever'].idxmax()]['year_release']

    return {"Año de lanzamiento con más horas jugadas para {}: {}".format(genero, max_year)}
```

## **2. Función UserForGenre (FUNCIONA. REVISADO)**


df_f2_f3 es resultado de realizar un segundo merge entre el resultado del paso 1 df_f1 y df_users_review_curado_sentiments utilizando las columnas user_id y item_id

def UserForGenre( genero : str ): Debe devolver el usuario que acumula más horas jugadas para el género dado y una lista de la acumulación de horas jugadas por año.
Ejemplo de retorno: {"Usuario con más horas jugadas para Género X" : us213ndjss09sdf, "Horas jugadas":[{Año: 2013, Horas: 203}, {Año: 2012, Horas: 100}, {Año: 2011, Horas: 23}]}

### **Codigo de la función** 

```python
def UserForGenre(genero: str):
    # Filtrar el DataFrame por el género específico
    df_genre = df_f2_f3[df_f2_f3['genres'] == genero]

    # Agrupar por usuario y año, sumar las horas jugadas
    grouped_df = df_genre.groupby(['user_id', 'year_release'])['playtime_forever'].sum().reset_index()

    # Encontrar al usuario con más horas jugadas
    max_user = grouped_df.loc[grouped_df['playtime_forever'].idxmax()]['user_id']

    # Filtrar el DataFrame original para obtener las horas jugadas por año del usuario máximo
    max_user_df = df_f2_f3[df_f2_f3['user_id'] == max_user]
    max_user_hours_by_year = max_user_df.groupby('year_release')['playtime_forever'].sum().reset_index()

    # Crear la lista de la acumulación de horas jugadas por año
    horas_jugadas_por_anio = [{"Año": int(anio), "Horas": int(horas)} for anio, horas in zip(max_user_hours_by_year['year_release'], max_user_hours_by_year['playtime_forever'])]

    # Crear el diccionario de retorno
    resultado = {"Usuario con más horas jugadas para Género " + genero: max_user, "Horas jugadas": horas_jugadas_por_anio}

    return resultado

# Llamada de ejemplo con el género 'Action'
resultado_ejemplo = UserForGenre('Casual')
print(resultado_ejemplo)

```

### Probar que la función es válida

```python
# Filtrar el DataFrame por el género específico
df_genre = df_f2_f3[df_f2_f3['genres'] == 'Casual']

# Agrupar por usuario y año, sumar las horas jugadas
grouped_df = df_genre.groupby(['user_id', 'year_release'])['playtime_forever'].sum().reset_index()

# Calcular la cantidad total de horas jugadas por género
total_hours_by_genre = df_genre.groupby('year_release')['playtime_forever'].sum().reset_index()

# Usuarios con más horas jugadas para el género 'Action'
top_users = grouped_df.groupby('user_id')['playtime_forever'].sum().reset_index().sort_values(by=['playtime_forever'], ascending=False)['user_id'].unique()[0:3]

# Crear un diccionario para almacenar los resultados
resultados_usuarios = {}

# Iterar sobre los usuarios seleccionados
for user_id in top_users:
    # Filtrar el DataFrame original para obtener las horas jugadas por año del usuario
    user_df = df_f2_f3[(df_f2_f3['user_id'] == user_id) & (df_f2_f3['genres'] == 'Casual')]
    user_hours_by_year = user_df.groupby('year_release')['playtime_forever'].sum().reset_index()

    # Crear la lista de la acumulación de horas jugadas por año
    horas_jugadas_por_anio = [{"Año": int(anio), "Horas": int(horas)} for anio, horas in zip(user_hours_by_year['year_release'], user_hours_by_year['playtime_forever'])]

    # Almacenar los resultados para este usuario
    resultados_usuarios[user_id] = {
        "Total horas jugadas": int(user_df['playtime_forever'].sum()),
        "Horas jugadas por año": horas_jugadas_por_anio
    }

# Mostrar los resultados
for user_id, info in resultados_usuarios.items():
    print(f"Usuario {user_id} - Total de horas jugadas: {info['Total horas jugadas']} - Horas jugadas por año para el género Action:")
    print(info["Horas jugadas por año"])
    print("\n")

# Mostrar la cantidad total de horas jugadas por género
print("Total de horas jugadas por género:")
print(total_hours_by_genre)
```

Al ver el ranking de usuarios se comprueba que la función opera como debería

## **3. Función users recommend** (FUNCIONA)

def UsersRecommend( año : int ): Devuelve el top 3 de juegos MÁS recomendados por usuarios para el año dado. (reviews.recommend = True y comentarios positivos/neutrales)
Ejemplo de retorno: [{"Puesto 1" : X}, {"Puesto 2" : Y},{"Puesto 3" : Z}]

Se utiliza el mismo merge de la funcion 2

### **Codigo de la función**

```python
def UsersRecommend(year: int):
    # Filtrar por año
    df_filtered = df_f2_f3[df_f2_f3['year_review'] == year]

    # Filtrar por recomendaciones positivas/neutrales
    df_filtered = df_filtered[(df_filtered['recommend'] == True) & (df_filtered['Sentiment_Score'].isin([1, 2]))]

    # Obtener el top 3 de juegos más recomendados
    top_games = df_filtered.groupby('item_name')['recommend'].count().sort_values(ascending=False).head(3)

    # Crear el diccionario de resultados en el formato deseado
    top_games_dict = {f'Puesto {i+1}': juego for i, (juego, count) in enumerate(zip(top_games.index, top_games.values))}

    # Devolver el diccionario en el formato deseado
    return top_games_dict

# Llamada de ejemplo con el año 2012
resultado_ejemplo1 = UsersRecommend(2012)
resultado_ejemplo1


```

### **Probar que la función es válida**

```python
# Convertir 'year_review' a tipo numérico (int) si no lo está ya
df_f2_f3['year_review'] = pd.to_numeric(df_f2_f3['year_review'], errors='coerce')

# Filtrar los años de interés
years_of_interest = [2012, 2013, 2014]
df_filtered_years = df_f2_f3[df_f2_f3['year_review'].isin(years_of_interest)]

# Filtrar solo los casos en los que 'recommend' es True y 'Sentiment_Score' es 1 o 2
filtered_data = df_filtered_years[(df_filtered_years['recommend'] == True) & (df_filtered_years['Sentiment_Score'].isin([1, 2]))]

# Agrupar por año, juego y calcular las estadísticas
grouped_stats = filtered_data.groupby(['year_review', 'item_name']).agg({
    'recommend': 'sum',
    'Sentiment_Score': 'sum'
}).reset_index()

# Ordenar por frecuencia de recomendaciones en orden descendente
sorted_stats = grouped_stats.sort_values(by=['year_review', 'recommend'], ascending=[True, False])

# Seleccionar los top 5 juegos para cada año
top5_per_year = sorted_stats.groupby('year_review').head(3)

print("Top 5 juegos recomendados por año:")
top5_per_year
```

Es válida. El orden que da la función es el mismo orden del ranking creado 

## **4. Función UsersWorstDeveloper**


def UsersWorstDeveloper( año : int ): Devuelve el top 3 de desarrolladoras con juegos MENOS recomendados por usuarios para el año dado. (reviews.recommend = False y comentarios negativos)
Ejemplo de retorno: [{"Puesto 1" : X}, {"Puesto 2" : Y},{"Puesto 3" : Z}]

Merge necesario

df_juegos, df_users_review_curado_sentiments, on='item_id', how='inner'

df_f4_f5


### **CODIGO DE LA FUNCION**

```python
def UsersWorstDeveloper(year: int):
    # Filtrar por el año deseado en la columna 'year_review'
    df_year = df_f4_f5[df_f4_f5['year_review'] == year]

    # Agrupar por desarrolladora y calcular las estadísticas
    grouped_stats = df_year.groupby('developer').agg({
        'recommend': lambda x: (x == False).sum(),
        'Sentiment_Score': lambda x: (x == 0).sum()
    }).reset_index()

    # Ordenar por la suma de juegos no recomendados en orden descendente
    sorted_stats = grouped_stats.sort_values(by=['recommend'], ascending=[False])

    # Seleccionar las top 3 desarrolladoras
    top3_developers = sorted_stats.head(3)

    # Crear el formato de retorno como un diccionario
    result = {f"Puesto {i}": developer for i, developer in enumerate(top3_developers['developer'], 1)}

    return result

# Ejemplo de uso
resultado_ejemplo1 = UsersWorstDeveloper(2013)
resultado_ejemplo1

```

### **Validación de la función**


```python
# Años de interés
years_of_interest = [2013, 2014, 2015]

# Mostrar resultados para cada año
for year in years_of_interest:
    # Filtrar por el año deseado en la columna 'year_review'
    df_year = df_f4_f5[df_f4_f5['year_review'] == year]

    # Agrupar por desarrolladora y calcular las estadísticas
    grouped_stats = df_year.groupby('developer').agg({
        'recommend': lambda x: (x == False).sum(),
        'Sentiment_Score': lambda x: (x == 0).sum()
    }).reset_index()

    # Ordenar por la suma de juegos no recomendados en orden descendente
    sorted_stats = grouped_stats.sort_values(by=['recommend'], ascending=[False])

    # Seleccionar las top 5 desarrolladoras
    top5_developers = sorted_stats.head(5)

    # Mostrar resultados para el año específico
    print(f"Resultados para el año {year}:")
    print(top5_developers)
    print("\n")

    print("\n")
```

Se observa que el resultado de la función, es decir el ordenamiento, es coherente tanto con la frecuencia de las no recomendaciones y como con la frencuencia de la evaluacion negativa de sentimiento.

### **CODIGO DE LA FUNCION**

```python
def sentiment_analysis(developer):
    # Filtrar el dataframe por la desarrolladora especificada
    developer_df = df_f4_f5[df_f4_f5['developer'] == developer]

    # Sumar los valores de Sentiment_Score para cada categoría
    negative_sum = developer_df[developer_df['Sentiment_Score'] == 0].shape[0]
    neutral_sum = developer_df[developer_df['Sentiment_Score'] == 1].shape[0]
    positive_sum = developer_df[developer_df['Sentiment_Score'] == 2].shape[0]

    # Crear el diccionario de salida
    result_dict = {
        developer: {
            'Negative': negative_sum,
            'Neutral': neutral_sum,
            'Positive': positive_sum
        }
    }

    return result_dict

# Ejemplo de uso
developer_name = "valve"
result = sentiment_analysis(developer_name)
result
```
