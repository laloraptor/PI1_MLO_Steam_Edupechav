## **Cargar base**

```python
import pandas as pd
import gzip
import ast

# Ruta al archivo comprimido gzip
archivo_gzip = "bases/users_items.json.gz"

# Lista para almacenar las filas del DataFrame
filas_usitems = []

# Abrir el archivo comprimido gzip y leer línea por línea
with gzip.open(archivo_gzip, 'rt', encoding='utf-8') as f:
    for line in f:
        # Evaluar cada línea y agregarla a la lista
        filas_usitems.append(ast.literal_eval(line))

# Crear un DataFrame a partir de la lista de filas
df_usit = pd.DataFrame(filas_usitems)

# Mostrar el DataFrame
print(df_usit)
```

Se elimina la columna user_url e items_count

```python
df_usit = df_usit.drop(["user_url", "items_count"], axis=1)

df_usit
```

Explorar valor nulo

```python
# Ver el tipo de datos de df_usit
tipos_datos = df_usit.dtypes

# Verificar la presencia de valores nulos
valores_nulos = df_usit.isnull().sum()

# Crear un nuevo DataFrame con la información
df_info = pd.DataFrame({
    "Tipos de Datos": tipos_datos,
    "Valores Nulos": valores_nulos
})

# Mostrar el DataFrame resultante
print(df_info)
```

desanidar

```python
# se normaliza la columna "items"
df_usit2 = pd.json_normalize(filas_usitems, record_path=["items"], meta=["steam_id","user_id"] )
df_usit=df_usit2
```

```python
# Definir el nuevo orden de las columnas
nuevo_orden = ["user_id", "steam_id", "item_id", "item_name", "playtime_forever", "playtime_2weeks"]

# Reorganizar las columnas del DataFrame
df_usit = df_usit[nuevo_orden]

# Mostrar el DataFrame resultante
df_usit
```

```python
# Verificar si hay valores nulos en todas las columnas
nulos_por_columna = df_usit.isnull().sum()

# Mostrar la frecuencia de nulos por columna
print("Frecuencia de valores nulos por columna:")
print(nulos_por_columna)
```

```python
# Verificar duplicados en df_usit
duplicados = df_usit.duplicated()

# Contar duplicados
cantidad_duplicados = duplicados.sum()


# Mostrar resultados
print("Cantidad total de registros duplicados en 'df_usit':", cantidad_duplicados)
```

eliminamos esos duplicados

```python
df_usit.drop_duplicates(inplace=True)
```
